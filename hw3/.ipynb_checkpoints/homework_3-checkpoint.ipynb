# Задача №2 - Реализовать Image-blending на основе сшивки по градиентам

Задача - взять фото двух лиц : ваше и друга, с помощью метода Poisson image editing совместить глаза, нос и рот с первого изображения со вторым. Суть в том, что при использовании такого метода границы совмещенного изображения не видны.

Статья, где описан метод  

Patrick Pérez, Michel Gangnet, and Andrew Blake. 2003. Poisson image editing. ACM Trans. Graph. 22, 3 (July 2003), 313–318. https://doi.org/10.1145/882262.882269

Пример такого совмещения:

<img src="../homework/blending/blending.png" alt="Drawing" style="width: 700px;"/>


```python
def laplacian_matrix(n, m):   
    mat_D = scipy.sparse.lil_matrix((m, m))
    mat_D.setdiag(-1, -1)
    mat_D.setdiag(4)
    mat_D.setdiag(-1, 1)
        
    mat_A = scipy.sparse.block_diag([mat_D] * n).tolil()
    
    mat_A.setdiag(-1, 1*m)
    mat_A.setdiag(-1, -1*m)
    
    return mat_A

def poisson(source, dest, mask):
    y_range, x_range = source.shape[:2]
    mat_a = laplacian_matrix(y_range, x_range)
    laplacian = mat_a.tocsc()

    for y in range(1, y_range - 1):
        for x in range(1, x_range - 1):
            if mask[y, x] == 0:
                k = x + y * x_range
                mat_a[k, k] = 1
                mat_a[k, k + 1] = 0
                mat_a[k, k - 1] = 0
                mat_a[k, k + x_range] = 0
                mat_a[k, k - x_range] = 0
    mat_a = mat_a.tocsc()
    mask_flat = mask.flatten()

    for channel in range(source.shape[2]):
        source_flat = source[:, :, channel].flatten()
        dest_flat = dest[:, :, channel].flatten()

        alpha = 1
        mat_b = laplacian.dot(source_flat)*alpha
        mat_b[mask_flat == 0] = dest_flat[mask_flat == 0]

        x = scipy.sparse.linalg.spsolve(mat_a, mat_b)
        x = x.reshape((y_range, x_range))
        x[x > 255] = 255
        x[x < 0] = 0
        x = x.astype('uint8')

        dest[:y_range, :x_range, channel] = x
    return dest
```

```python
obj = cv2.imread('blending/obj.png')
background = cv2.imread('blending/background.jpg')
mask = cv2.imread('blending/mask.png', cv2.IMREAD_GRAYSCALE) > 0

out = poisson(obj, background, mask)
plt.imshow(out)
plt.show()
```

# Задача №3 - Найди клетки

Даны снимки раковых клеток. Существует задача - определить стадию рака клетки по такому изображению. Для того, чтобы подойти к решению классификации рака клетки, необходимо сначала подготовить данные.

Исходные изображения в реальных задачах могут быть очень большого размера (более 20000 px). Однако из визуального анализа можно заметить, что большая часть этих снимков пустая и не несет в себе полезную информацию.

Ваша задача выделить небольшие ячейки изображений из исходного так, чтобы на ячейках было только изображение клетки.

В качестве аргументов у функции будут значения:

```python
img = cv2.imread('cells/train0_1.jpeg')
img = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
def find_cells(img, sz, count):
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    mask = (img_gray < 200).astype(np.uint8)
    
    indent = sz // 2
    kernel = np.ones((sz, sz), dtype=np.uint8)

    corr = cv2.filter2D(mask, cv2.CV_16S, kernel, borderType=cv2.BORDER_CONSTANT)
    overlap_map = corr > 0.5 * sz * sz
    overlap_map[:sz] = 0
    overlap_map[-indent:] = 0
    overlap_map[:, :indent] = 0
    overlap_map[:, -indent:] = 0

    parts = []
    coords = []
    h, w, _ = img.shape
    
    for i in range(h):
        row = overlap_map[i]
        if np.sum(row) != 0:
            gap = []
            j = 0
            while j < w:
                if row[j] != 0:
                    idxs = [j]
                    while row[j] > 0 and j < w:
                        j += 1
                    idxs.append(j)
                    gap.append(idxs.copy())
                j += 1
                
            for sf in gap:
                start, end = sf
                amount_cells = (end - start) // cell_size
                point = (end - start) // 2 + start
                if amount_cells <= 1:
                    parts.append(img[i - indent: i - indent + sz, point - indent: point - indent + sz])
                    coords.append([[i - indent, point - indent], [i - indent + sz, point - indent + sz]])
                else:
                    array_indent = (sz * amount_cells) // 2
                    point -= array_indent
                    for k in range(amount_cells):
                        parts.append(img[i - indent: i - indent + sz, point + k * sz: point + (k+1)*sz])
                        coords.append([[i - indent, point + k * cell_size], [i - indent + cell_size, point + (k+1)*cell_size]])
            overlap_map[i - sz: i + sz] = 0
            
    arr_h = ceil(np.sqrt(count))
    arr_w = ceil(count / arr_h)

    out = np.zeros((arr_h * sz, arr_w * sz, 3), dtype=np.uint8)
    for i, cut in enumerate(parts[:count]):
        h = i // arr_w
        w = i % arr_w
        out[h * sz: (h + 1) * sz, w * sz: (w+1)*sz] = cut
        
    return out, coords
```

```python
out, coords = find_cells(img, 200, 40)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(out)

for pt1, pt2 in coords:
    cv2.rectangle(img, (pt1[1], pt1[0]), (pt2[1], pt2[0]), (255, 0, 0), 10)
ax2.imshow(img)
```
